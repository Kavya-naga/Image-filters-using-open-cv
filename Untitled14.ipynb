```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

pixels = float(10)
img = cv2.imread('your_uploaded_image.jpg')  # Replace with your actual image filename

# Flip image horizontally
img = cv2.flip(img, 1)

# Create brightness variations
img_1 = img + pixels
img_1[img_1 < 0] = 0
img_1[img_1 > 255] = 255
img_1 = img_1.astype(np.uint8)

img_2 = img + (2 * pixels)
img_2[img_2 < 0] = 0
img_2[img_2 > 255] = 255
img_2 = img_2.astype(np.uint8)

img_3 = img + (3 * pixels)
img_3[img_3 < 0] = 0
img_3[img_3 > 255] = 255
img_3 = img_3.astype(np.uint8)

# Display images
cv2_imshow(img)        # Original
cv2_imshow(img_1)      # Filter-1
cv2_imshow(img_2)      # Filter-2
cv2_imshow(img_3)      # Filter-3
```
