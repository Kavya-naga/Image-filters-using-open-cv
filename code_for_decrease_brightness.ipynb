```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
pixels = float(10)

img = cv2.imread('kavya photo.jpg')
img = cv2.flip(img, 1)

img_1 = img - pixels
img_1[img_1 <  0 ] = 0
img_1[img_1 > 255] = 255
img_1 = img_1.astype(np.uint8)

img_2 = img - (2*pixels)
img_2[img_2 <  0 ] = 0
img_2[img_2 > 255] = 255
img_2 = img_2.astype(np.uint8)

img_3 = img - (3*pixels)
img_3[img_3 <  0 ] = 0
img_3[img_3 > 255] = 255
img_3 = img_3.astype(np.uint8)


cv2_imshow(img)
cv2_imshow(img_1)
cv2_imshow(img_2)
cv2_imshow(img_3)


```
