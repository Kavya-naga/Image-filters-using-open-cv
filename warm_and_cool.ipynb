```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
def add_tone(img_path, color, a, b):
    background = []
    img = cv2.imread(img_path)
    cols = img.shape[1]
    rows = img.shape[0]
    for i in range(rows):
        temp = []
        for j in range(cols):
            temp.append(color)
        background.append(temp)
    background = np.array(background).astype(np.uint8)
    final = cv2.addWeighted(img, a , background, b , 0)
    cv2_imshow(img)
    cv2_imshow(final)
add_tone('kavya photo.jpg', [80,20,200], .5, .5)
```
